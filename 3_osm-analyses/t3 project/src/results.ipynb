```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
"""Reading resulting tables for Indonesia"""

import pandas as pd
import matplotlib.pyplot as plt

time_list = ["pre", "post"]
fig = plt.figure(dpi=200)

print("Statistics for Indonesia\n----------")
for time in time_list:
    df = pd.read_excel(f"../results/ind_{time}_pop_per_building.xlsx")
    mean = df[f"osm_pop_{time}"].mean()
    print(f"Mean {time}: {mean}")
    df[f"osm_pop_{time}"].plot.hist(bins=1000, legend=True, xlim=(0, 25), alpha=0.33)
mean_ems = df[f"ems_pop"].mean()
print(f"Mean EMS: {mean_ems}")
df["ems_pop"].plot.hist(bins=1000, legend=True, xlim=(0, 25), alpha=0.33)
plt.title("OSM population per building in Indonesia")
plt.show()
```
