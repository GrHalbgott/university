**Research question 1 - Calculate the betweenness centrality based on the shortest route and the one based on fastest route using the networkx package. Where are the biggest differences between them? Which one do you think is more realistic? Create figures to show your findings.**



```python
import warnings
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# ignore FutureWarnings from geopandas
warnings.simplefilter(action="ignore", category=FutureWarning)

img_path = Path("../output")
fastest_path = img_path / "Karlsruhe_nx_fastest" / "Karlsruhe_nx_fastest.gpkg"
shortest_path = img_path / "Karlsruhe_nx_shortest" / "Karlsruhe_nx_shortest.gpkg"
```

```python
fastest = gpd.read_file(fastest_path)
shortest = gpd.read_file(shortest_path)
```

```python
# Histogram comparison of centrality

sns.histplot(data=fastest, x="centrality", kde=True, binwidth=0.0005, alpha=0.5, label="Fastest")
sns.histplot(data=shortest, x="centrality", kde=True, binwidth=0.0005, alpha=0.5, label="Shortest")
plt.xlim(0, 0.01)
plt.title("Centrality comparison between shortest and fastest routes")
plt.legend()
plt.show()
```

```python
# Histogram comparison of centrality

sns.histplot(data=fastest, x="centrality", kde=True, binwidth=0.0005, alpha=0.5, label="Fastest")
sns.histplot(data=shortest, x="centrality", kde=True, binwidth=0.0005, alpha=0.5, label="Shortest")
plt.xlim(0.05, 0.12)
plt.ylim(0, 11)
plt.title("Centrality comparison between shortest and fastest routes")
plt.legend()
plt.show()
```

```python
import warnings
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# ignore FutureWarnings from geopandas
warnings.simplefilter(action="ignore", category=FutureWarning)

img_path = Path("../output")
nopop_path = img_path / "Karlsruhe_gibc_1000" / "Karlsruhe_gibc_1000.gpkg"
pop_path = img_path / "Karlsruhe_gibc_1000_pop" / "Karlsruhe_gibc_1000_pop.gpkg"
```

```python
pop = gpd.read_file(pop_path)
nopop = gpd.read_file(nopop_path)
```

```python
# Histogram comparison of centrality

sns.histplot(data=pop, x="centrality", kde=True, binwidth=1, alpha=0.5, label="With PDBRRG")
sns.histplot(data=nopop, x="centrality", kde=True, binwidth=1, alpha=0.5, label="No PDBRRG")
plt.xlim(0, 30)
plt.title("Centrality comparison between GIBC with and without PDBRRG")
plt.legend()
plt.show()
```

```python
# Histogram comparison of centrality

sns.histplot(data=nopop, x="centrality", kde=True, binwidth=1, alpha=0.5, label="No PDBRRG")
sns.histplot(data=pop, x="centrality", kde=True, binwidth=1, alpha=0.5, label="With PDBRRG")
plt.xlim(50, 155)
plt.ylim(0, 20)
plt.title("Centrality comparison between GIBC with and without PDBRRG")
plt.legend()
plt.show()
```

```python
# Histogram comparison of centrality

sns.histplot(data=nopop, x=len("geometry"), kde=True, binwidth=1, alpha=0.5, label="No PDBRRG")
sns.histplot(data=pop, x=len("geometry"), kde=True, binwidth=1, alpha=0.5, label="With PDBRRG")
plt.xlim(50, 155)
plt.ylim(0, 20)
plt.title("Centrality comparison between GIBC with and without PDBRRG")
plt.legend()
plt.show()
```
